```python
import pandas as pd 
import numpy as np
from sklearn.preprocessing import StandardScaler as ss
from sklearn.preprocessing import MinMaxScaler as mms
from sklearn.preprocessing import LabelEncoder as le
```

```python
def cleandataset(a):
    a=a.drop_duplicates()
    
    clmns=a.columns
    nullv=a.isnull().sum()
    datat=a.dtypes
    
    for z in range(len(clmns)):
        print("Would you like to drop the column "+ clmns[z]+ " ?[y/n]")
        w=input()
        if w=='y'or w=='Y':
                a=a.drop([clmns[z]], axis=1)
                print("Column dropped")
                
    clmns=a.columns
    nullv=a.isnull().sum()
    datat=a.dtypes
          
    
    print(" For string enter 1")
    print(" For integer enter 2")
    print(" For float enter 3")
    print("Include dates and contact-num in string")
    
    
    for y in range(len(clmns)):
        print("Column"+ str(y)+ " is "+ clmns[y])
        print("Enter type of column")
        y1=int(input())
        if y1==1:
                for k in range(len(a)):
                    try:
                        a[clmns[y]][k]=int(a[clmns[y]][k])
                        a[clmns[y]][k]=np.nan
                    except:
                         a[clmns[y]][k]= a[clmns[y]][k]
        elif y1==2:
            a[clmns[y]]=pd.to_numeric(a[clmns[y]], errors='coerce')
            
        elif y1==3:
            a[clmns[y]]=pd.to_numeric(a[clmns[y]], errors='coerce')
            
            
    
    datat=a.dtypes
    nullv=a.isnull().sum()
    clmns=a.columns
    for j in range(len(clmns)):
        
        if nullv[j]!=0:
            print("Found "+str(nullv[j])+" empty values in column "+ str(clmns[j]))
            print("Would you like to drop the column?[y/n]")
            s=input()
            if s=='y' or s=='Y':
                a=a.drop([clmns[j]], axis=1)
                print("Column dropped")
            else:
                if datat[j]=='int64' or datat[j]=='float64':
                    print("Would you like to fillup null cells with mean value?[y/n] ")
                    g=input()
                    s=clmns[j]
                    #v=a[[s]].mean()
                    if g=='y' or g=='Y':
                        a[[s]]=a[[s]].fillna(a[[s]].mean())
                        print("Operation Successful")
                    else:
                        print("Would you like to fillup null cells with median value?[y/n] ")
                        g=input()
                        if g=='y' or g=='Y':
                            a[[clmns[j]]]=a[[clmns[j]]].fillna(a[clmns[j]].median())
                            print("Operation Successful")
                        else:
                            print("Would you like to fillup null cells with any specific value?[y/n] ")
                            g=input()
                            if g=='y' or g=='Y':
                                print("Enter the value")
                                if datat[j]=='int64':
                                    d=int(input())
                                else:
                                    d=float(input())
                                a[[clmns[j]]]=a[[clmns[j]]].fillna(value=d)
                                print("Operation Successful")
                
                else:
                            print("Would you like to fillup null cells with any specific value?[y/n] ")
                            g=input()
                            if g=='y' or g=='Y':
                                print("The row is:")
                            
                                '''for u in range(len(clmns)):
                                    g=clmns[u][j]
                                    k.append(g)'''
                            
                                print("Enter the value")
                                d=input()
                                print("Enter the value")
                                a[[clmns[j]]]=a[[clmns[j]]].fillna(value=d)
                                print("Operation Successful")
                           # else:
                            #    print("Would you like to remove the row?[y/n]")
                             #   g=input()
                              #      if g=='y' or g=='Y':
                               #         d=input()
                                #        a[[clmns[j]]]=a[[clmns[j]]].fillna(value=d)
                                 #       print("Operation Successful")
    

    print("Data cleaning completed")
    clmns=a.columns
    nullv=a.isnull().sum()
    datat=a.dtypes
    print("Data preprocessing ")
    print(" For setting dummies enter 1")
    print(" For label encoding(categorical) enter 2")
    print("For no actions enter 3")    
    for z1 in range(len(datat)):
            if datat[z1]=='object':
                print(clmns[z1] + " has " +str(len(a[clmns[z1]].unique()))+ " different values")
            
                print("Which technique you would like to use for "+ clmns[z1]+ " ?[1/2/3]")
                zi=int(input())
                if zi==1:
                    df2=pd.get_dummies(a[[clmns[z1]]], drop_first=True)
                    a=pd.concat([a,df2], axis=1)
                    a=a.drop([clmns[z1]],axis=1)
                elif zi==2:
                    s=le()
                    a[[clmns[z1]]]=s.fit_transform(a[[clmns[z1]]])
                else:
                    print("No action for this input")
            

    
    clmns=a.columns
    datat=a.dtypes
    nullv=a.isnull().sum()
    print("Would you like to remove outliers?[y/n]")
    h=input()
    if h=='y' or h=='Y':
        for h in range(len(datat)):
            
                if datat[h]=='int64' or datat[h]=='float64':
                    print("Working on column "+ clmns[h])
                    n=clmns[h]
                    q1=a[n].quantile(0.25)
                    q3=a[n].quantile(0.75)
                    iqr=q3-q1
                    ll=q1-1.5*iqr
                    ul=q3+1.5*iqr
                    a=a[(a[n]>ll) & (a[n]<ul)]
    a=a.dropna()
    
    
    
    
    print("Would you like to scale your data?[y/n]")
    f=input()
    if f=='y' or f=='Y':
        
        clmns=a.columns
        nullv=a.isnull().sum()
        datat=a.dtypes


        for dt in range(len(datat)):
                if datat[dt]=='int64' or datat[dt]=='float64':
                    print(clmns[dt])

                    print("For Min-Max-Scalerization enter 1")

                    print("For Standard Scalerization enter 2")
                    fi=input()

                    if fi=='1':
                            mmso=mms(feature_range=(0,1))
                            a[[clmns[dt]]]=mmso.fit_transform(a[[clmns[dt]]])

                    elif fi=='2':
                            so=ss()
                            a[[clmns[dt]]]=so.fit_transform(a[[clmns[dt]]])

    print("Done")
    return a   
```
